```python
import pandas as pd
import os
import zipfile
from datetime import datetime
import re
```

```python
with zipfile.ZipFile(r'D:\Tennis-Project\data\raw\tennis_data.zip', 'r') as zip_ref:
    name = zip_ref.namelist()[0]
    zip_ref.extract(name, path='../Datas/')
```

```python
with zipfile.ZipFile(r'D:\Tennis-Project\Datas\20240205.zip', 'r') as zip_ref:
    zip_ref.extractall(path='../Datas/extracted')
```

```python
for path in os.listdir(r'D:\Tennis-Project\Datas\extracted\data\raw'):
    full_path = os.path.join('../Datas/extracted/data/raw', path)
    
    cats = set()
    for filename in os.listdir(full_path):
        file_name = re.sub(r'_\d+\.parquet', '', filename)
        cats.add(file_name)
        
    print(path, cats)
```
