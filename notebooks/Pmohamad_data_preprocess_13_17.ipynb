```python
import pandas as pd
import numpy as np
```

**Q.13** **********************************

```python
df_away = pd.read_parquet('../data/raw/parquets/match_away_team.parquet', engine = 'pyarrow')
df_home = pd.read_parquet('../data/raw/parquets/match_home_team.parquet', engine = 'pyarrow')

df_away.head()
```

```python
df_home.head()
```

```python
print(len(df_home), len(df_away))
```

```python
Q_13_data = pd.concat([
    df_away[['player_id', 'plays']],
    df_home[['player_id', 'plays']]
]).drop_duplicates()
```

```python
Q_13_data.head()
```

```python
print(len(Q_13_data))
```

```python
Q_13_data.to_parquet('../data/raw/parquets/Q_13_data.parquet')
```

**Q.15** *********************************

```python
df_venue = pd.read_parquet('../data/raw/parquets/match_venue.parquet', engine = 'pyarrow')
df_venue.head()
```

```python
distinct_countries = set(
                     list(df_away['country'])+
                     list(df_home['country'])+
                     list(df_venue['country']) )
Q_15_data = pd.DataFrame(list(distinct_countries), columns = ['countries'])
print(Q_15_data.head())
print(len(Q_15_data))
```

```python
Q_15_data.to_parquet('../data/raw/parquets/Q_15_data.parquet')
```
